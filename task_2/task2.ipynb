```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Load Dataset, I used iris dataset
df = pd.read_csv("IRIS.csv") 
print("Shape of dataset:", df.shape)
```

```python
print("\nFirst 5 rows:")
print(df.head())
```

```python
#summary statistics
print("\nSummary Statistics:")
print(df.describe(include='all'))

print("\nInfo:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
#histogram for numerical plots
df.hist(figsize=(10, 6), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numerical Features")
plt.tight_layout()
plt.show()
```

```python
#boxplot
plt.figure(figsize=(12, 6))
for idx, column in enumerate(df.select_dtypes(include=np.number).columns, 1):
    plt.subplot(1, 4, idx)
    sns.boxplot(y=column, data=df, color='lightgreen')
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()
```

```python
#pairplot
sns.pairplot(df, hue='species', diag_kind='kde', palette='husl')
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()
```

```python
#correlation matrix, heatmap plot
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
#scatter plot, it is an interactive plot
fig = px.scatter(df, x="sepal_length", y="sepal_width", color="species",
                 size="petal_length", hover_data=["petal_width"],
                 title="Interactive Sepal Scatter Plot")
fig.show()
```

```python
#3d scatter plot
fig = px.scatter_3d(df, x='sepal_length', y='sepal_width', z='petal_length',
                    color='species', size='petal_width', symbol='species',
                    title="3D Scatter Plot of Iris Dataset")
fig.show()
```

```python
#counting the species
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='species', palette='Set2')
plt.title("Count of Each Species")
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(12, 6))
for idx, col in enumerate(df.select_dtypes(include=np.number).columns, 1):
    plt.subplot(2, 2, idx)
    sns.violinplot(data=df, x='species', y=col, palette='Pastel1')
    plt.title(f'Violin plot of {col} by Species')
plt.tight_layout()
plt.show()
```

```python
#basic inference
print("Petal length and width are good indicators for separating species.")
print("Setosa is clearly separable based on petal size.")
print("Sepal dimensions show some overlap among species.")
print("No missing values; data is clean.")
```
